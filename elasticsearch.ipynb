```python
from pprint import pprint
from elasticsearch import Elasticsearch

es = Elasticsearch('http://localhost:9200')
client_info = es.info()
print("Connected to Elasticsearch version %s" % client_info['version']['number'])
pprint(client_info)
```

```python
es.indices.create(index='first_index', ignore=400, settings={
    "index": {
        "number_of_shards": 1,
        "number_of_replicas": 0
    }})
# es.indices.
```

```python
document = {
    'title':"ArithmeticError",
    'text':"text",
    "created_on": "2017-01-01"
}

response = es.index(index='first_index', body=document)
response
```

```python
import requests
url = 'https://dummyjson.com/products'
json_response = requests.get(url)
pprint(json_response.json())
```

```python
def insert_document(document):
    response = es.index(index='first_index', body=document)
    return response
```

```python
def print_info(response):
    print(f"Document ID: {response['_id']} is {response['result']} and split into {response['_shards']['successful']} shards")
```

```python
json = json_response.json()
for docs in json["products"]:
    response = insert_document(docs)
    print_info(response)
```

```python
index_mapping = es.indices.get_mapping(index='first_index')
pprint(index_mapping)
```

```python
def print_info(response):
    """
    Pretty prints an Elasticsearch ObjectApiResponse
    """
    import json
    
    if hasattr(response, 'body'):
        # If it's an ObjectApiResponse, use its body
        print(json.dumps(response.body, indent=2))
    else:
        # In case it's already a dictionary or another format
        print(json.dumps(response, indent=2))

print_info(index_mapping)
```
